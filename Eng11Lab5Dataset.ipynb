```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt


datain = pd.read_csv('/content/drive/MyDrive/2. Sophomore/Spring24/Engin11/E11/indoordata.csv')

#import datadatain = pd.read_csv('/content/drive/MyDrive/2. Sophomore/Spring24/Engin11/E11/indoordata.csv')




datain.values
```

```python
#import data
dataout = pd.read_csv('/content/drive/MyDrive/2. Sophomore/Spring24/Engin11/E11/outdoordata.csv')

dataout.values
dataout.values
datain.values
diff = (dataout["Time"]-datain["Time"])
a = diff[0]
a


```

```python
particle_columns = ['pm10 standard', 'pm25 standard', 'pm100 env']

for column in particle_columns:
    plt.figure(figsize=(10, 5))

    plt.plot(datain['Time']+a, datain[column], label='Inside', color='blue')


    plt.plot(dataout['Time'], dataout[column], label='Outside', color='red')

    plt.xlabel('Time')
    plt.ylabel(column)
    plt.title(f'{column} Comparison')
    plt.legend()
    plt.show()
```

```python
import math
#hist together
plt.figure(figsize=(10, 6))
plt.hist(datain['pm25 standard'], alpha=0.5, label='Inside PM25 Standard', color='blue')
plt.hist(dataout['pm25 standard'], alpha=0.5, label='Outside PM10', color='red')

plt.xlabel('Data')
plt.ylabel('Frequency')
plt.title('Inside and Outside Data')
plt.legend()
plt.show()

#means and stds
inside_mean = datain['pm25 standard'].mean()
inside_std = datain['pm25 standard'].std()
outside_mean = dataout['pm25 standard'].mean()
outside_std = dataout['pm25 standard'].std()

# separation
std_mean = inside_std/(math.sqrt(len(datain['pm25 standard'])))
mean_diff = abs(inside_mean - outside_mean)

if mean_diff > 3 * std_mean:
    print("The means are separated by more than 3sigma of the uncertainty on the mean.")
    print("There is only a .3% probability that the two sets of data are measuring the same true distribution.")
else:
    print("The means are not separated by more than 3sigma of the uncertainty on the mean.")
```
